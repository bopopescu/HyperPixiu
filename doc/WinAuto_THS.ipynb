```python
# focus on the window with title=u'网上股票交易系统5.0', and get rid of some pop-up news et al.
def resetFocus() :
    for w in app.windows(class_name='#32770') :
        # print(w.WindowText)
        if w.window_text() != u'网上股票交易系统5.0':
            w.close()
    mainWin = app.top_window()
    mainWin.set_focus()
    pywinauto.keyboard.SendKeys('{F4}') # stay at list page
    
    return mainWin
    
```

```python
# buy version 1
def sendBuy(security, price, amount) :
    tw = resetFocus()
    pywinauto.keyboard.SendKeys('{F1}')

    try:
        tw.window_(title_re=u'重填', control_type='Button').click()
    except:
        pass# tw.print_control_identifiers()
    
    strPrice = '%.3f' %price
    strAmount = '%d' %amount

    pywinauto.keyboard.SendKeys('{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}'+ security) # enter 证券代码
    pywinauto.keyboard.SendKeys('{TAB}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}' + strPrice) # enter 买入价格
    pywinauto.keyboard.SendKeys('{TAB}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}{BACK}' + strAmount) # enter 买入数量
    pywinauto.keyboard.SendKeys('{TAB}{ENTER}') # 买入按钮

    orderId=""
    cfmtext=''
    try :
        time.sleep(0.1)
        cfmw= app.top_window()
        # cfmw.print_control_identifiers()
    
        cfmtext=cfmw.window_(title_re = u'股东帐号：', control_type='Static').window_text()
    except:
        pass
    
    # print(cfmtext) # u'股东帐号：AXXXXXXXXX  证券代码：600237 买入价格：3.600  买入数量：300  预估金额：1086.000  您是否确定以上买入委托？'
    syntax =u'^.*\n证券代码：'+security+'.*\n买入价格：'+strPrice +'.*\n买入数量：'+strAmount+'\n.*'
    # syntax =u'.*'+security+'.*'+price +'.*'+amount+'.*'
    cfmexpress = re.compile(syntax)
    if cfmexpress.match(cfmtext) :
        print("order confirmed")
        pywinauto.keyboard.SendKeys('%y')

        try :
            time.sleep(0.5)
            cfmw= app.top_window()
            cfmtext=cfmw.window_(title_re = u'.*成功提交', control_type='Static').window_text() # 您的买入委托已成功提交，合同编号：76466795。
            syntax=u'.*成功提交.*合同编号：([^。]*).*'
            m = re.compile(syntax).match(cfmtext)
            print(m.groups())
            btn=cfmw.window_(title_re = u'确定')
            btn.click()
        
            orderId= m.groups()[0]
        except:
            pass
    else :
        print("order mismatch, cancel it")
        pywinauto.keyboard.SendKeys('%n')
    
    resetFocus()
    return orderId
```

```python
sendBuy(security='600237', price = 4.060, amount = 300)
```
